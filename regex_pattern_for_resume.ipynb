# **Regex-pattern**

```python
# Global variable for the image path
image_path = r"/content/resu_new.jpg"

def extract_text_from_image():
      # Open the image using PIL
      image = Image.open(image_path)

      # Use pytesseract to extract text from the image
      extracted_text = pytesseract.image_to_string(image)
      text = " ".join(extracted_text.split('\n'))

      return text

def extract_name_from_resume(text):
    name = None

    # Use regex pattern to find a potential name
    pattern = r"(?<![\w\d])([A-Z][a-zA-Z]{1,}(?:\s[A-Z][a-zA-Z]{1,}){0,2})(?![\w\d])"
    match = re.search(pattern, text)
    if match:
        name = match.group()

    return name

def extract_designation_from_resume(text):
    designation = None

    # Use regex pattern to find a potential designation
    pattern = r"\b(?:Software\sEngineer|Web\sDeveloper|Data\sScientist|Project\sManager|Product\sManager|UX\sDesigner|UI\sDesigner|Business\sAnalyst|Systems\sAnalyst|Database\sAdministrator|Network\sAdministrator|DevOps\sEngineer|Front-End\sDeveloper|Back-End\sDeveloper|Full-Stack\sDeveloper|QA\sEngineer|Technical\sLead|IT\sManager|Systems\sEngineer|Application\sDeveloper|Cloud\sEngineer|Security\sAnalyst|Site\sReliability\sEngineer|Solutions\sArchitect|Consultant|Intern|Junior\sDeveloper|Senior\sDeveloper|Lead\sDeveloper|Technical\sArchitect|Program\sManager|Technical\sSupport|Support\sEngineer|Operations\sManager|Digital\sMarketer|Data\sAnalyst|Chief\sTechnology\sOfficer|Chief\sInformation\sOfficer|Chief\sExecutive\sOfficer|Chief\sMarketing\sOfficer)\b"
    matches = re.findall(pattern, text,re.IGNORECASE)
    if matches:
        designation = matches[0]

    return designation

def extract_contact_number_from_resume(text):
    contact_number = None

    # Use regex pattern to find a potential contact number
    pattern = r"\+?\d{1,3}[-.\s]?\(?\d{2,4}\)?[-.\s]?\d{3,4}[-.\s]?\d{4,6}\b"
    match = re.search(pattern, text)

    if match:
        contact_number = match.group()

    return contact_number

def extract_email_from_resume(text):
    email = None

    # Use re.findall() to find all potential email addresses in the text
    pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    matches = re.findall(pattern, text)

    # Return the first found email address
    if matches:
        email = matches[0]

    return email

def extract_linkedin_from_resume(text):
    linkedin = None

    # Use re.findall() to find all potential linkedin addresses in the text
    pattern = r'\b(?:https?:\/\/)?(?:www\.)?linkedin\.com\/(?:in\/[\w-]+|pub\/[\w-]+|company\/[\w-]+|school\/[\w-]+|groups\/[\w-]+|events\/[\w-]+|[\w-]+(?:\/[\w-]*)?)\b'
    matches = re.findall(pattern, text,re.IGNORECASE)

    # Return the first found linkedin address
    if matches:
        linkedin = matches[0]

    return linkedin

def extract_location_from_resume(text):
    location = None

    # Use re.findall() to find all potential location addresses in the text
    pattern = r'\b[A-Z][a-zA-Z]*(?: [A-Z][a-zA-Z]*)?, [A-Z]{2}\b|\b[A-Z][a-zA-Z]*, [A-Z]{2}\b'
    matches = re.findall(pattern, text,re.IGNORECASE)

    # Return the first found linkedin address
    if matches:
        location = matches[0]

    return location


if __name__ == '__main__':
    text = extract_text_from_image()

    name = extract_name_from_resume(text)
    if name:
        print("Name:", name)
    else:
        print("Name not found")

    designation = extract_designation_from_resume(text)
    if designation:
        print("Designation:", designation)
    else:
        print("designation not found")

    contact_number = extract_contact_number_from_resume(text)
    if contact_number:
        print("Contact Number:", contact_number)
    else:
        print("Contact Number not found")

    email = extract_email_from_resume(text)
    if email:
        print("Email:", email)
    else:
        print("Email not found")

    linkedin = extract_linkedin_from_resume(text)
    if linkedin:
        print("Linkedin:", linkedin)
    else:
        print("Linkedin not found")

    location = extract_location_from_resume(text)
    if location:
        print("Location:", location)
    else:
        print("location not found")

```
